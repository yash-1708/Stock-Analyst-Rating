```python
driver = webdriver.Chrome("chromedriver.exe")
driver.get("https://www.nseindia.com/products/content/equities/equities/eq_security.htm")
```

```python

for i in range(1,20):
    sym = df1.at[i, 'company_name']#to get the company name from the df
    dte = df1.at[i, 'predict_date']#to get the date from the df
    dte= dte + relativedelta(years=1)#adding 1 year to the current date
    dte1=dte + relativedelta(days=7)
    dte1=dte1.strftime('%d-%m-%Y')
    dte2=dte - relativedelta(days=7)
    dte2=dte2.strftime('%d-%m-%Y')
    symbol = driver.find_element_by_id('symbol')
    rangRadioBtn= driver.find_element_by_id('rdDateToDate')
    fromDate = driver.find_element_by_id('fromDate')
    toDate= driver.find_element_by_id('toDate')
    submitBtn= driver.find_element_by_id('get')
    symbol.send_keys(sym)#Symbol Name
    rangRadioBtn.click();
    fromDate.send_keys(dte2)#date should be in this fromat only
    toDate.send_keys(dte1)
    submitBtn.click();
    try:
        element = WebDriverWait(driver, 20).until(
        EC.presence_of_element_located((By.XPATH, "//div[@id='historicalData']/table")))
            
        tableData = driver.find_element_by_xpath("//div[@id='historicalData']/table")
        temp = tableData.text
        tableclean=temp.split('\n')
        size=len(tableclean)
        closeval = []
        for i in range(6,size):
            temp2 = tableclean[i]
            tabclean2 = temp2.split(" ")
            requiredval=tabclean2[8]
            closeval.append(requiredval)
        dfclose = pd.DataFrame({'closeval':closeval})
        #dfnumeric = pd.to_numeric(dfclose['closeval'], downcast='float')
        #mean = dfnumeric.mean()
        print(sym)
        print(dfclose)
        #print(mean)
    except TimeoutException as e:
        print ("error", sym)
    except IndexError as ie:
        print ("error", sym)
    finally :
        print("clear form done")
        symbol.clear()#Clear Symbol Name
        fromDate.clear()#date should be in this fromat only
        toDate.clear()
print("tadaaaaaa")
```
